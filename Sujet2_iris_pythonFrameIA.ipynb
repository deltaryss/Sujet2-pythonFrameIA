```python
import pandas as pd
import matplotlib.pyplot as plt
#import seaborn as sns
```

```python
data = pd.read_csv('iris.csv')
data.head()
```

```python
data.columns
```

```python
data.describe()
```

```python
color_dict = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
data['target'] = data['target'].map(color_dict)
```

```python
#show the dispersion of the data with matplotlib with the color of the points according to the target
plt.scatter(data['sepal length (cm)'], data['sepal width (cm)'], c=data['target'])
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()

#sns.scatterplot(data=data, x="sepal length (cm)", y="sepal width (cm)", hue="target")
```

```python
# Plot the similar graphic for petal width vs. petal length:
plt.scatter(data['petal length (cm)'], data['petal width (cm)'], c=data['target'])
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```
